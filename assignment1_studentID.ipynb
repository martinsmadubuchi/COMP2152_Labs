```python
# Section c: Create workout_stats dictionary
# Dictionary with friend names as keys and tuples of workout minutes as values
workout_stats = {
    "Alex": (30, 45, 20),  # (yoga, running, weightlifting)
    "Jamie": (45, 30, 50),  # (yoga, running, weightlifting)
    "Taylor": (20, 60, 40)  # (yoga, running, weightlifting)
}
```

```python
# Section d: Calculate total workout minutes for each friend using a loop
for friend, workout_minutes in workout_stats.items():
    total_minutes = sum(workout_minutes)
    workout_stats[f"{friend}_Total"] = total_minutes

print("Updated workout_stats dictionary:")
print(workout_stats)
```

```python
# Section e: Create 2D list from dictionary
# List of lists containing workout minutes for each friend
workout_list = []
for friend in ["Alex", "Jamie", "Taylor"]:
    workout_list.append(list(workout_stats[friend]))

print("\nWorkout list (2D list):")
print(workout_list)
```

```python
# Section f: Slice the workout_list
# Extract yoga and running minutes for all friends (columns 0 and 1)
print("\nYoga and Running minutes for all friends:")
for i, friend_name in enumerate(["Alex", "Jamie", "Taylor"]):
    print(f"{friend_name}: Yoga={workout_list[i][0]}, Running={workout_list[i][1]}")

# Extract weightlifting minutes for the last two friends
print("\nWeightlifting minutes for the last two friends:")
for i, friend_name in enumerate(["Jamie", "Taylor"]):
    print(f"{friend_name}: Weightlifting={workout_list[i+1][2]}")
```

```python
# Section g: Check if any friend's total workout minutes >= 120
print("\nChecking for friends with 120+ minutes of workout:")
for friend in ["Alex", "Jamie", "Taylor"]:
    total_key = f"{friend}_Total"
    if total_key in workout_stats and workout_stats[total_key] >= 120:
        print(f"Great job staying active, {friend}!")
```

```python
# Section h: User input feature to search for friend
friend_name = input("Enter a friend's name: ")

if friend_name in workout_stats and not friend_name.endswith("_Total"):
    workout_minutes = workout_stats[friend_name]
    total_minutes = workout_stats[f"{friend_name}_Total"]
    print(f"\n{friend_name}'s workout minutes:")
    print(f"Yoga: {workout_minutes[0]} minutes")
    print(f"Running: {workout_minutes[1]} minutes")
    print(f"Weightlifting: {workout_minutes[2]} minutes")
    print(f"Total workout minutes: {total_minutes} minutes")
else:
    print(f"Friend {friend_name} not found in the records.")
```

```python
# Section i: Print friend with highest and lowest total workout minutes
friends_list = ["Alex", "Jamie", "Taylor"]
totals = {friend: workout_stats[f"{friend}_Total"] for friend in friends_list}

max_friend = max(totals, key=totals.get)
min_friend = min(totals, key=totals.get)

print(f"\nFriend with the highest total workout minutes: {max_friend} ({totals[max_friend]} minutes)")
print(f"Friend with the lowest total workout minutes: {min_friend} ({totals[min_friend]} minutes)")
```
